```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2 as cv
import csv
import os
```

```python
def img_load(path):
    img = cv.imread(path)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)    
    return img 
```

```python
def img_load_gray(path):
    img = cv.imread(path, 0)
    return img 
```

```python
def img_show(img):
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
def img_threshold(img):
    ret, thresh = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
    return thresh
```

```python
def img_resize(img):
    img = cv.resize(img, (28, 28))
    return img
```

```python
def img_to_list(img):
    img_list = list()
    for i in img.flatten():
        img_list.append(i)    
    return img_list
```

```python
def img_process(path):
    img = img_load_gray(path)
#     img = img_resize(img)
    img = img_threshold(img)
    return img
```

```python
img = img_load('result/2/i10.png')
img = img_threshold(img)
img_show(img)
```

```python
img = img_process('result/2/i10.png')
img_show(img)
```

```python
all_img_list = list()
for i in range(10):
    for j in range(200):
        try:
            img = img_process('result/%d/i%d.png' % (i, j))
            img_list = list()
            img_list.append(i)            
            img_list += img_to_list(img)
            all_img_list.append(img_list)
        except:
            print('result/%d/i%d.png' % (i, j))
```

```python
all_img_array = np.asarray(all_img_list)
print(all_img_array.shape)
```

```python
path = 'dataset/400_400.csv'
with open(path, 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile)
    index_list = [i for i in range(785)]
    spamwriter.writerow(index_list)
    for i in all_img_array:
        spamwriter.writerow(i)
```
