```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
def img_load(path):
    img = cv.imread(path)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)    
    return img 
```

```python
def img_show(img):
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
def img_compare(img0, img1):
    fig = plt.gcf()
    fig.set_size_inches(10, 20)
    ax0 = plt.subplot(1, 2, 1)
    ax0.imshow(img0)
    ax0.set_xticks([]), ax0.set_yticks([])
    ax1 = plt.subplot(1, 2, 2)
    ax1.imshow(img1)
    ax1.set_xticks([]), ax1.set_yticks([])
    plt.show()
```

```python
for i in range(10):
    origin_path = 'channel3_size32/x_train/%s.png' % str(i)
    halftone_path = 'halftone/x_train/%s.png' % str(i)
    img0, img1 = img_load(origin_path), img_load(halftone_path)
    img_compare(img0, img1)
```
