```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
import os
from PIL import Image
```

```python
def img_load(path):
    img = cv.imread(path)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)    
    return img
```

```python
def img_resize(img, size):
    img = cv.resize(img, (size, size))
    return img
```

```python
def img_show(img):
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
def img_save(img, path):
    img = Image.fromarray(img)
    img.save(path, 'png')    
```

```python
img = img_load('channel3_size32/x_train/0.png')
img_show(img)
```

```python
img1 = img_resize(img, 256)
img_show(img1)
img_save(img1, 'channel3_size256/test2')
```

```python
path = 'channel3_size32/x_train/'
path_to = 'channel3_size256/x_train/'
for i in range(60000):
    i = str(i) + '.png'
    print(i)
    img = img_load(path + i)
    img = img_resize(img, 256)        
    img_save(img, (path_to + i))
```
