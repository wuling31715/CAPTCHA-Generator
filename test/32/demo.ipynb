```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
import os
```

```python
def img_load(path):
    img = cv.imread(path)
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)    
    return img
```

```python
def img_show(img):
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
for i in range(10):
    img = img_load('demo/%d.png' % i)
    img_show(img)
```
